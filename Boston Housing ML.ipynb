# Model Building

```python
## Independent and Dependent features

X=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1]
```

```python
X.head()
```

```python
##Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
y
```

```python
# Internally We use gradient descent
# Main aim is to come near the global minima
# All features are in different units
# If we had same units in all the independent variables, the converging would be faster
# Therefore, we standardize the dataset

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler() # < this is the standard scaling object
```

```python
# Pass the my x train dataset and transform it completely
X_train=scaler.fit_transform(X_train)
```

```python
# In test set, we dont write fit_transform
# Because we are gonna make sure that whatever the info we have w.r.t training dataset
# And whatever the techniques we have applied for transforming it, the same needs to be applied to test dataset
# This is done to make sure that our model does not know much about the test dataset

X_test = scaler.transform(X_test)
```

```python
# no need to do this for the output
```

# Training the model (Linear Regression Algorithm)

```python
# importing linear regression model
from sklearn.linear_model import LinearRegression
```

```python
# initialize the specific model
regression = LinearRegression()
```

```python
# train the model
regression.fit(X_train, y_train) # we are creating a hyperplane here cuz we have so many features here
```

```python
# print the coefficients and the intercept
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
# finding out on which parameters that this model has been trained
regression.get_params()
```

```python
# doing predictions with test data
reg_pred = regression.predict(X_test)
```

```python
reg_pred # see all the predictions
```

```python
# model adequacy checking
```

```python
## plot a scatter plot for the prediction
plt.scatter(y_test,reg_pred)
```

You can see that the above plotting is linear. When this is the case, we can say that our model has actually performed well.

```python
# calculating residuals (error)
residuals = y_test - reg_pred
```

```python
residuals
```

```python
# plotting these residuals
sns.displot(residuals, kind = "kde")
```

Our main assumption is to get the above curve as a normal distribution
Here residuals vary between -10 to +10
Some points are ranging from 10-30
But overall this looks good

```python
## Scatter plot with respect to prediction and residuals
## uniform distribution
plt.scatter(reg_pred,residuals)
```

Above is the residuals vs fitted values plot
You can see that majority of the points are scattered uniformly
This looks good and it gives the indication that our model is doing well.

To be so sure about this, we have to use Performance Metrics

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(y_test, reg_pred))
print(mean_squared_error(y_test, reg_pred))

# root mse
print(np.sqrt(mean_squared_error(y_test, reg_pred)))
```

Formula

R^2 = 1 - SSR/SST

R^2 = coefficient of determination SSR = sum of squares of residuals SST = total sum of squares

```python
# Another performace metric
# R2 and Adjusted R2 Square

from sklearn.metrics import r2_score
score=r2_score(y_test,reg_pred)
print(score)
```

Adjusted R2 = 1  [(1-R2)*(n-1)/(n-k-1)]

where:

R2: The R2 of the model n: The number of observations k: The number of predictor variables

# Pickling the model file for development

Pickle is a python module used to serialize a python object into a binary format and deserialize it back to the python object.
We pickle the regression model file so that we can further use it for deployment.
In other words, it's the process of converting a Python object into a byte stream to store it in a file/database, maintain program state across sessions, or transport data over the network.
The pickle module keeps track of the objects it has already serialized, so that later references to the same object won't be serialized again, thus allowing for faster execution time. Allows saving model in very little time.
https://towardsdatascience.com/why-turn-into-a-pickle-b45163007dac


```python
import pickle
```

```python
# wb - write byte mode
pickle.dump(regression, open('regmodel.pkl', 'wb'))
```

```python
# loading the pickle
pickled_model=pickle.load(open('regmodel.pkl','rb'))
```

```python
## Prediction
pickled_model.predict(scaler.transform(boston.data[0].reshape(1,-1)))
```
